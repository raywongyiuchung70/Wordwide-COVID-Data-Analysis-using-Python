## **Activity 2:**

Data Cleaning and Feature Engineering Enhancing data relevance and structure, enabling focused analysis on key COVID-19 indicators and demographic factors.

1.  Impute missing values in the dataset columns

2.  Remove duplicate rows from the DataFrame.

3.  Create new features if needed (e.g., extract year and month from the 'date' column).

4.  Explore unique countries in the dataset and count the total number of countries.

```python
# Impute missing values in the dataset columns
# iterates through columns with numerical data types (float64 and int64)
# and replaces missing values (NaN) with the median value of that column.
for col in df.select_dtypes(include=['float64','int64']).columns:
    df[col].fillna(df[col].median(), inplace=True)

# Fill categorical (object type) columns with mode
# iterates through columns with object data types (categorical) and
# replaces missing values with the mode (most frequent value) of that column.
for col in df.select_dtypes(include=['object']).columns:
        df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
# 2. Remove duplicate rows
df.drop_duplicates(inplace=True)
```

```python
# 3. Feature engineering: add year and month
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
```

```python
# 4. Explore unique countries
countries = df['location'].unique()
print("Total unique countries:", len(countries))
print(countries)
```

## **Activity 3:**

Worldwide COVID-19 Overview Providing a global and temporal perspective on COVID-19, identifying patterns and correlations crucial for strategic decision-making.
1. Visualize the WHO Regions (Continent column in dataset) with total COVID-19 cases and total deaths by using bar plots. .

```python
import matplotlib.pyplot as plt
import seaborn as sns


# 1. WHO Regions (continents) - total cases and deaths
region_df = df.groupby('continent')[['total_cases','total_deaths']].max().reset_index()

print("Region-level summary:\n", region_df)

plt.figure(figsize=(10,6))
region_df.plot(
    x='continent',
    y=['total_cases','total_deaths'],
    kind='bar',
    figsize=(10,6)
)
plt.title("Total COVID-19 Cases and Deaths by Continent")
plt.ylabel("Count")
plt.xlabel("Continent")
plt.show()
```

## **Insights:**

*	Highest total COVID-19 cases & deaths cases both appeared on Africa
*	Asia, which was the 1st COVID-19 cases reported was only have the 3rd total cases and 4th total death cases.
*	North America surprisingly ranked 3rd on the Total cases & 2nd total death cases across continent despite their reputations on science/technology & health care capabilities


```python
# 2. Worldwide monthly trend of cases
monthly_cases = df.groupby(['year','month'])['new_cases'].sum().reset_index()
monthly_cases['df'] = pd.to_datetime(monthly_cases[['year','month']].assign(day=1))

print("Monthly cases trend sample:\n", monthly_cases.head())

plt.figure(figsize=(12,6))
plt.plot(monthly_cases['df'], monthly_cases['new_cases'])
plt.title("Worldwide Monthly COVID-19 Cases")
plt.ylabel("New Cases")
plt.xlabel("Date")
plt.show()
```

## **Insight**:
* 2020 – Early Spread
 Steady rise in global cases as the virus spread worldwide.
  * Multiple moderate waves, peaking late 2020 to early 2021.
* 2021–2022 – Major Surges
  * Early 2022 saw the largest global spike (~400M cases in a month), driven by the Omicron variant.
  * Another significant surge occurred in late 2022 (~360M cases).
* 2023 – Decline & Stabilization
  * Sharp fall in case numbers after Omicron waves.
  * By mid-2023, global cases stabilized at relatively low levels.
* Overall Pattern
  * Surges were variant-driven (e.g., Omicron), with steep rises and rapid declines.
  * By 2023, widespread vaccinations, natural immunity, and treatments helped contain outbreaks


```python
# 3. Correlation between total cases and deaths
corr = df[['total_cases','total_deaths']].corr()
print("Correlation:\n", corr)

plt.figure(figsize=(5,4))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation between Total Cases and Deaths")
plt.show()

```

##**Insight** :

* The graph shows that COVID-19 total cases and deaths are very strongly linked (correlation ~0.94), but not perfectly.

* This means while more cases usually result in more deaths, healthcare, vaccines, and variant severity significantly influence outcomes.

* Monitoring the divergence between cases and deaths over time can help assess the effectiveness of interventions.

* For example: A drop in correlation over time would signal improvements in survival rates. A rise in correlation could signal stress on healthcare systems or more severe variants.

## **Activity 5:**

Time Series Analysis Capturing the daily dynamics of COVID-19, assessing the impact of vaccination, and monitoring testing metrics for a comprehensive global overview.

1.  Explore the daily trend of new cases and new deaths globally using line plots.

2.  Calculate and visualise the daily average of new cases and deaths globally.

3.  Explore the trend of vaccination coverage over time globally.

4.  Analyze the total tests and positive rate over time globally.

```python
# 1 Explore the daily trend of new cases and new deaths globally using line plots.
plt.figure(figsize=(12,6))
daily_global = df.groupby('date')[['new_cases', 'new_deaths']].sum().reset_index()
sns.lineplot(data=daily_global, x='date', y='new_cases', label='New Cases')

sns.lineplot(data=daily_global, x='date', y='new_deaths', label='New Deaths')
plt.title("Daily Trend of New Cases and Deaths")
plt.xlabel("Date")
plt.show()
```

## **Key Insights**
* Early Outbreak (2020)

  * Low case numbers in early 2020, followed by steady increases from mid-2020.
  * Deaths were visible but at a relatively low scale compared to cases.
  * First Major Waves (2020–2021)
  * Multiple moderate surges in new cases during late 2020 and early 2021.
  * Deaths rose alongside cases, though the scale of deaths remained much smaller compared to infections.

* Delta Wave (Mid 2021)
  * Noticeable spikes in mid-2021, linked to the Delta variant, which was more severe.
  * Deaths slightly increased but remained lower than expected given the sharp rise in cases — likely reflecting improvements in healthcare and early vaccinations.

* Omicron Wave (Late 2021 – Early 2022)
  * The largest surge in daily new cases is seen in early 2022, peaking close to 40 million daily cases.
  * Despite the massive rise in cases, deaths did not rise proportionally, confirming Omicron’s high transmissibility but lower lethality compared to Delta.

* Late Pandemic (Mid 2022 – 2023)
  * Multiple smaller spikes are seen after the Omicron peak, but none reached the same magnitude.
  * New cases steadily declined into 2023.
  * Deaths remained consistently low despite case spikes — showing effectiveness of vaccines, immunity, and treatments.

* End of 2023 – Stabilization
  * Both new cases and deaths approach very low levels by late 2023.
  * Indicates that the pandemic’s global impact had largely subsided.


```python
# 2 Calculate and visualise the daily average of new cases and deaths globally.
daily_avg = df.groupby('date')[['new_cases', 'new_deaths']].mean().reset_index()
sns.lineplot(data=daily_avg, x='date', y='new_cases', label='New Cases')

sns.lineplot(data=daily_avg, x='date', y='new_deaths', label='New Deaths')
plt.title("Daily Trend of New Cases and Deaths")
plt.xlabel("Date")
plt.show()
```

# **Insights from Comparing the Daily Average New Cases & Deaths with the  Raw Daily New Cases & Deaths**

* Similar but smoother trend compare to the previous Raw Daily New Cases & Deaths graph.

* Peaks and troughs are less jagged compared to raw daily data because the graph aggregates values (averaging reduces daily reporting noise).

* The average daily case graph removes volatility shows the daily average trend, eliminate short-term fluctuations and highlights major pandemic waves (2020, Delta in 2021, Omicron in 2022).

* The average daily case graphs makes easier to spot long-term patterns and major waves without noise.

* Better for trend analysis, while the previous raw daily new case & death case graph is better for real-time daily tracking.



```python
# 3 Explore the trend of vaccination coverage over time globally.

daily_vacc = df.groupby('date')['people_vaccinated_per_hundred'].sum().reset_index()
plt.figure(figsize=(12,6))
sns.lineplot(data=daily_vacc, x='date', y='people_vaccinated_per_hundred', label='People Vaccinated')
plt.title("Vaccination Coverage Over Time")
plt.xlabel("Date")
plt.ylabel("People Vaccinated")
plt.show()
```

* The fatality rate (deaths ÷ cases) was extremely high in early 2020, peaking above 200% at one point.

* This anomaly happened because testing was very limited early on, so only severe cases were being detected. Reported deaths often exceeded reported cases in the same time frame.

**2.	Mid to Late 2020 – Decline**

* As testing expanded, the fatality rate dropped sharply.
o	Still, the rate remained relatively elevated (~20–40%), reflecting limited treatment options and overwhelmed healthcare systems in many countries.

**3.	2021 – Delta Variant Period**

* During the Delta wave, the fatality rate showed slight fluctuations but did not spike dramatically.

* This suggests that while Delta caused severe illness and many deaths, it was balanced by much higher case detection through broader testing.

**4.	2022 – Omicron Surge**

* The fatality rate declined further, stabilizing at very low levels (~1–5%).

* Omicron was far more transmissible but less deadly, combined with widespread vaccinations and natural immunity, leading to a much lower death-to-case ratio.

**5.	2023 – Stabilization at Low Fatality**

* By 2023, the fatality rate remained consistently low.

* This shows that COVID-19 had shifted into a more endemic pattern, with better treatment availability, vaccines, and population immunity preventing high fatality levels.




```python
# 2 Explore the positivity rate (total_cases/ total tests) versus total tests conducted to analyse testing
# effectiveness using the x-axis as the logarithmic scale for better visualisation.
df['positivity_rate'] = df['total_cases']/ df['total_tests']
sns.scatterplot(data=df, x='date', y='positivity_rate')
plt.title("Positivity Rate Over Time")
plt.xlabel("Date")
plt.ylabel("Positivity Rate")
plt.xticks(rotation=30)
plt.show()
```

**2.	Mid-2020 – Decline in Fatality Rate**

* As testing ramped up globally, more mild and moderate cases were detected. This brought the fatality rate down significantly, stabilizing around lower levels (~20–40%).

**3.	2021 (Delta Period)**

* Some fluctuations are visible around 2021, aligning with the Delta variant surge.

* While Delta was more severe, the higher volume of detected infections kept the fatality rate from spiking dramatically.

**4.	2022 (Omicron Surge)**

* Fatality rate dropped further during Omicron waves, staying at very low levels (~1–5%).

* Omicron was far more transmissible but less deadly, and vaccines plus improved treatments helped reduce deaths relative to infections.
